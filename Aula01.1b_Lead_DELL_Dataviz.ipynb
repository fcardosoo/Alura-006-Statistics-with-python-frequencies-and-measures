```python
import pandas as pd
```

```python
# Os dados deste dataset são referentes ao consumo de álcool por pessoa em diferentes países,
#  coletados pela Organização Mundial da Saúde no ano de 2010.

df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/alcohol-consumption/drinks.csv')
```

```python
df.head()
```

```python
# consultando o Brasil
df[df['country'] == 'Brazil']
```

```python
# Mostrando os países com volume consumido de Alcool maior de 12 litros
df[df['total_litres_of_pure_alcohol']>=12].describe()
```

```python
# Mostrando as primeiras 5 linhas dos países com consumo ZERO de Álcool
df[df['total_litres_of_pure_alcohol'] == 0].head()
```

```python
# Ordenando os dados com o 'sort_values'
df.sort_values(by='wine_servings', ascending=False).head()
```
